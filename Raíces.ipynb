```python
from scipy import optimize
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return (x**5-6*x**4+2*x**3+20*x**2-27*x+10)
```

```python
x = np.linspace(-10, 10, 400)
y = f(x)
plt.plot(x, y, label='y = x^5-6x^4+2x^3+20x^2-27x+10')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la función y = x^5-6x^4+2x^3+20x^2-27x+10')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fprime=lambda x: 5 * x**4 - 24 * x**3 + 6 * x **2 + 40 * x -27
rng = np.random.default_rng()
x = rng.standard_normal(10)
a = np.arange(-5, 5)
vec_res = optimize.newton(f, x, fprime, maxiter=200)
loop_res = [optimize.newton(f, x0, fprime,
                            maxiter=200)
            for x0, a0 in zip(x, a)]
np.allclose(vec_res, loop_res)
analytical_result = np.sign(a) * np.abs(a)**(1/3)
fig, ax = plt.subplots()
ax.plot(a, analytical_result, 'o')
ax.set_xlabel('$a$')
ax.set_ylabel('$x$ where $f(x, a)=0$')
plt.show()
```
